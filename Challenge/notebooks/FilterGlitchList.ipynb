<h1>Filter Glitch list</h1>

```python
import os
import numpy as np
import pandas as pd
```

```python
# Read file
list_filename="gspy-db-20180813.csv"
# O1 dates: 2015 Sep 12th 0:00 UTC (GPS 1126051217) to 2016 Jan 19 16:00 UTC (GPS 1137254417)
```

```python
data_dir = os.path.join(os.path.dirname(os.getcwd()),"data")
```

```python
gl_df = pd.read_csv(os.path.join(data_dir,list_filename))
```

```python
gl_df.describe()
```

```python
gl_df.sample(10)
```

```python
# get times
# O1 dates: 2015 Sep 12th 0:00 UTC (GPS 1126051217) to 2016 Jan 19 16:00 UTC (GPS 1137254417)
t_max = np.max(gl_df.GPStime)
t_min = np.min(gl_df.GPStime)
print("T from GPS %.3f to %.2f (dt=%.3f , %.3f days)" % (t_min,t_max,t_max-t_min,(t_max-t_min)/86400.))
```

## Select O1 Glitches

```python
#Select glitches above a threshold in confidence and with a label
c_thr = 0.0#9 #95
snr_thr = 7.5
tmin_sel = 1126051217
tmax_sel = 1137254417
gl_df_sel = gl_df[(gl_df["GPStime"]>tmin_sel) & (gl_df["GPStime"]<tmax_sel) & (gl_df["snr"]>snr_thr) & (gl_df["confidence"]>c_thr) & (gl_df["label"].str.len()>1)]
```

```python
print("selected %d out of %d glitches" % (len(gl_df_sel),len(gl_df)))
```

```python
gl_df_sel.describe()
```

```python
gl_df_sel = gl_df_sel.drop(columns=["chisq","chisqDof","imgUrl","confidence","amplitude"])
```

```python
gl_df_sel.sample(6)
```

```python
out_col = list(gl_df.columns)
out_filename = os.path.join(data_dir,list_filename[:-4]+"_O1_filtered_t"+str(int(t_min))+"-"+str(int(t_max))+"_snr"+str(snr_thr)+".csv")
print(out_filename)
gl_df_sel.to_csv(out_filename,columns=out_col,index_label="Index")
```

### Select only the GravitySpy Glitches and filter out the time chunks not available from GWOSC

```python
# Read the gspy public table
gspy_train_df = pd.read_csv(os.path.join(data_dir,"trainingset_v1d1_metadata.csv"))
gspy_train_df.sample(5)
```

```python
# Iterate and select only the glitches in the gravityspy list
target_id_list=np.asarray(gspy_train_df["gravityspy_id"])
print(len(target_id_list))
print(target_id_list)
```

```python
sel_id_index=[]
mask_selected=np.full(len(gl_df_sel),False)

# print(mask_selected)
for gi in range(len(gl_df_sel)):
    src_id=gl_df_sel.iloc[gi]["id"]
    if src_id in target_id_list:
        #print(src_id)
        sel_id_index.append(src_id)
        mask_selected[gi]=True
        
print(len(sel_id_index))
print(mask_selected)
```

```python
gl_df_sel_gspy = gl_df_sel[mask_selected]
len(gl_df_sel_gspy)
```

```python
# save to a file
out_filename = os.path.join(data_dir,list_filename[:-4]+"_O1_filtered_t"+str(int(t_min))+"-"+str(int(t_max))+"_snr"+str(snr_thr)+"_gspy.csv")
print(out_filename)
gl_df_sel_gspy.to_csv(out_filename,index=False)#index_label="Index")
```

```python
# Now filter out the glitches not accessible to GWOSC
remove_df = pd.read_csv(os.path.join(data_dir,"glitch_to_remove_logfile_v1.csv"),sep="\t")
print(len(remove_df))
```

```python
remove_df.sample(3)
```

```python
remove_id_list = np.asarray(remove_df["id"])
mask_remove=np.full(len(gl_df_sel_gspy),True)
print(len(mask_remove))

# print(mask_selected)
for gi in range(len(gl_df_sel_gspy)):
    src_id=gl_df_sel_gspy.iloc[gi]["id"]
    #print(src_id+"-")
    if src_id in remove_id_list:
        print("Remove "+src_id)
        #sel_id_index.append(src_id)
        mask_remove[gi]=False
        
# print(len(sel_id_index))
# print(mask_remove)
```

```python
gl_df_sel_gspy_rem = gl_df_sel_gspy[mask_remove]
# gl_df_sel_gspy_rem=gl_df_sel_gspy_rem.drop(columns=["confidence","amplitude"])

print(len(remove_df))

print(len(gl_df_sel_gspy))
print(len(gl_df_sel_gspy_rem))
```

```python
src_dir="/home/razzano/razzanoWork/VirgoWork/GlitchData/2019_braga_school_O1_data_v1"
target_dir="/home/razzano/razzanoWork/VirgoWork/GlitchData/2019_braga_school_O1_data_out_v1"

import glob
import shutil

gl_df_sel_gspy_rem_id=np.asarray(gl_df_sel_gspy_rem["id"])
missing_mask=np.full(len(gl_df_sel_gspy_rem),False)
selected_final_mask=np.full(len(gl_df_sel_gspy_rem),False)

for gii in range(len(gl_df_sel_gspy_rem_id)):
    src_id = gl_df_sel_gspy_rem_id[gii]
    
    # search for glitch
    # print(src_id)
    src_file = glob.glob(os.path.join(src_dir,"*"+src_id+"*.hdf5"))
    if len(src_file)>0:
        print("copy",src_file)
        selected_final_mask[gii]=True
        shutil.copy(src_file[0],target_dir)
    else:
        print("missing ",src_id)
        missing_mask[gii]=True
```
